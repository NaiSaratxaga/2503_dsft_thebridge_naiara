### Step 1. This is the data given as a dictionary. Create a dataframe and assign it to a variable called army. 

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, None, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, None, 3, None, 2, None, 2, None, None, 1, 2, None],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

```python
army = pd.DataFrame(raw_data)
army
```

- regiment: Indica el nombre del regimiento al que pertenece cada unidad del ejército.
- company: Representa el número de la compañía dentro de cada regimiento. En los ejércitos suelen dividirse en varias compañías (por ejemplo, la 1ª o 2ª compañía) dentro de un regimiento.
- deaths: Indica el número de muertes que ocurrieron en cada unidad del regimiento o compañía.
- battles: El número de batallas en las que ha participado cada unidad.
- size: Representa el tamaño de la unidad, es decir, el número de soldados en cada compañía.
- veterans: El número de veteranos (soldados con experiencia) en cada unidad.
- readiness: Un valor que indica el nivel de preparación o disponibilidad de la unidad. Este valor podría representar el estado de alerta, la disposición para el combate, etc. Si el valor es None, significa que no se tiene información sobre este parámetro.
- armored: Indica si la unidad tiene vehículos blindados. El valor es binario (0 o 1), donde 1 significa que tienen vehículos blindados y 0 significa que no.
- deserters: El número de soldados que han desertado de la unidad.
- origin: La ubicación o el estado de origen del regimiento o unidad.

### Step 3. Drop features, if there are more than 30% missing values.

### Step 4. Fill missing values with the mean of their regiment.

```python
 # army.groupby("regiment"): Agrupa el DataFrame army por la columna regiment
 # Después de hacer el groupby(), seleccionamos únicamente la columna deaths.
 # Así que para cada grupo (que corresponde a un valor único de regiment), trabajamos solo con los valores de la columna deaths.
 # La función transform() aplica una operación a cada grupo y devuelve una nueva serie del mismo tamaño que la original.
 # Aquí es donde estamos transformando los valores de cada grupo de la columna deaths de manera individual.
 # lambda val: val.fillna(val.mean()): val: Este es el argumento que representa cada uno de los grupos dentro de la columna deaths.
 # Es una serie de valores correspondientes a un grupo específico de regiment.
 # val.mean(): Calcula la media de los valores de deaths en ese grupo específico de regiment.
 # val.fillna(val.mean()): Luego, para cada grupo, reemplaza todos los valores NaN de la serie con la media de ese grupo.
 # fillna() es un método de pandas que llena los valores faltantes (NaN) con un valor específico, en este caso la media calculada del grupo.


army["deaths"] = army.groupby("regiment")["deaths"].transform(lambda val: val.fillna(val.mean()))
army.info()
army["deaths"]
```

```python
# Aquí compruebo que el que tenía el valos NaN ha sido reemplazado por la media
army
```

### Step 5. Set the 'origin' colum as the index of the dataframe

```python
# Tenemos que establecer la columna origin como índice del df.
# El método set_index() asignará la columna 'origin' como índice del DataFrame, lo que significa que las filas estarán indexadas por los valores de esa columna.

army.set_index("origin", inplace=True)

army.info()

```

### Step 6. Select the 'deaths', 'size' and 'deserters' columns from Maine and Alaska

```python
# Utilizamos .loc[] para seleccionar las filas correspondientes a "Maine" y "Alaska"
# y las columnas "deaths", "size" y "deserters" de esas filas.
select_deaths_size_deserters = army.loc[["Maine", "Alaska"], ["deaths", "size", "deserters"]]

select_deaths_size_deserters
```

### Step 7. Select the rows 3 to 7 and the columns 3 to 6

```python
# Utilizamos .iloc[] para seleccionar las filas 3 a 7 (inclusive) y las columnas 3 a 6 (inclusive).
# .iloc[] permite seleccionar por posiciones enteras, no por etiquetas.

select_rows_columns = army.iloc[3:8, 3:7]

select_rows_columns
```

### Step 8. Select every row after the fourth row and all columns

```python
# El primer parámetro especifica las filas desde la 4 hasta el final, y el segundo parámetro : indica todas las columnas.
every_row = army.iloc[4:, :]  # Seleccionamos desde la fila 4 (inclusive) hasta el final.

every_row
```

### Step 9. Select every row up to the 4th row and all columns

```python
# Seleccionamos desde la fila 0 hasta la fila 3, todas las columnas con (:)
select_every_row = army.iloc[:4, :]  

select_every_row
```

### Step 10. Select the 3rd column up to the 7th column

```python
# Seleccionamos desde la columna con índice 2 hasta la columna con índice 6 
select_columns_3_to_7 = army.iloc[:, 2:7]  

select_columns_3_to_7
```

### Step 11. Select rows where df.deaths is greater than 50

```python
# Usamos .loc[] para seleccionar las filas donde 'deaths' es mayor que 50
select_deaths_50 = army.loc[army["deaths"] > 50]

select_deaths_50
```

### Step 12. Select rows where df.deaths is greater than 500 or less than 50

```python
# Usamos .loc[] para seleccionar las filas donde 'deaths' sea mayor que 500 o menor que 50
# Tenemos que crear 2 condiciones y usamos el operador OR
select_deaths_500_or_50 = army.loc[(army["deaths"] > 500) | (army["deaths"] < 50)]

select_deaths_500_or_50
```

### Step 13. Select all the regiments not named "Dragoons"

```python
# Usamos .loc[] para seleccionar todas las filas donde la columna 'regiment' no sea 'Dragoons'
select_not_dragoons = army.loc[army["regiment"] != "Dragoons"]

select_not_dragoons
```

### Step 14. Select the rows called Texas and Arizona

```python
# Usamos .loc[] para seleccionar las filas donde el índice ('origin') sea 'Texas' o 'Arizona'
# Usamos isin() para verificar si el índice está en la lista ["Texas", "Arizona"]

select_texas_arizona = army.loc[army.index.isin(["Texas", "Arizona"])]

select_texas_arizona

```

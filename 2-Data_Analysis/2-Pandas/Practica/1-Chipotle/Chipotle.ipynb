### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
# podemos usar sort_values y iloc
# Encontramos el artículo más caro ordenando los precios de los productos de mayor a menor.
# Seleccionamos el primero de la lista (el artículo más caro).
# Accedemos a la cantidad que fue pedida de ese artículo.

# Ordenamos el DataFrame por 'item_price' de mayor a menor
most_expensive_item = chipo.sort_values(by="item_price", ascending=False).iloc[0]

# Obtenemos la cantidad de ese artículo
quantity_most_expensive = most_expensive_item["quantity"]

# Mostrar el resultado
print(quantity_most_expensive)

# explicación:
# sort_values(by="item_price", ascending=False): Ordena el DataFrame chipo de manera descendente 
# (de mayor a menor) según la columna item_price.
# .iloc[0]: Selecciona la primera fila, que corresponde al artículo más caro.
# most_expensive_item["quantity"]: Accede a la columna quantity del artículo más caro y muestra cuántos se han pedido de ese artículo.
```

```python
chipo["unit_price"].max()
```

```python
chipo[chipo["unit_price"] == chipo ["unit_price"].max()]["quantity"].sum()
```

```python
# Encontramos el índice del artículo con el precio más alto
idx_max_price = chipo["item_price"].idxmax()

# Obtenemos la cantidad de ese artículo
quantity_most_expensive = chipo.loc[idx_max_price, "quantity"]

# Mostrar el resultado
print(quantity_most_expensive)

# explicacion
# chipo["item_price"].idxmax(): Encuentra el índice de la fila donde item_price tiene el valor máximo (es decir, el precio más alto).
# chipo.loc[idx_max_price, "quantity"]: Usa el índice encontrado para acceder a la columna quantity
# en esa fila específica y obtener la cantidad de ese artículo.
```

```python
chipo[chipo["unit_price"] == chipo ["unit_price"].max()]["quantity"].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
# Filtrar pedidos donde se ha pedido más de una "Canned Soda"
more_than_one_soda = chipo.query('item_name == "Canned Soda" and quantity > 1')

# Contar cuántos pedidos cumplen con esta condición
count_soda_orders = more_than_one_soda.shape[0]

# Mostrar el resultado
print(count_soda_orders)

# query('item_name == "Canned Soda" and quantity > 1'): Filtra las filas del DataFrame
# en las que el nombre del artículo es "Canned Soda" y la cantidad es mayor a 1.
#shape[0]: Obtiene el número de filas (que corresponde a cuántos pedidos cumplen con esa condición).
```

```python
# Conectamos con la base de datos chinook.db
connection = sqlite3.connect(r"data/chinook.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# 1 Obtén los clientes de brasil
query = '''
SELECT * FROM customers
WHERE customers.Country = "Brazil";
'''

df = sql_query(query)
df
```

```python
# 2  Obtén los empleados que son agentes de ventas
query = '''
SELECT * FROM employees
WHERE title = "Sales Support Agent"
'''

df = sql_query(query)
df
```

```python
# 3 Obtén las canciones de ‘AC/DC’

query = '''
SELECT * FROM tracks 
WHERE Composer = "AC/DC";
'''

df = sql_query(query)
df
```

```python
# 5 Obtén los empleados que son agentes de ventas: Nombre completo, Dirección (Ciudad, Estado, País) y email

query = '''
SELECT E.FirstName, E.LastName, E.ADDRESS, E.CITY, E.State, E.Country, E.EMAIL FROM EMPLOYEES AS E
WHERE E.Title = "Sales Support Agent";
'''

df = sql_query(query)
df
```

```python
# 6 Obtén una lista con los países no repetidos a los que se han emitido facturas

query = '''
SELECT I.BillingCountry FROM INVOICES AS I
GROUP BY I.BILLINGCOUNTRY
'''

df = sql_query(query)
df
```

```python
# 11. Cuántas facturas hay de cada país

query = '''
SELECT BillingCountry, COUNT(*) AS Facturas  
FROM invoices  
GROUP BY BillingCountry
ORDER BY facturas desc;
'''

df = sql_query(query)
df
```

```python
#  12. Cuántas facturas ha habido en 2009 y 2011

query = '''
SELECT strftime("%Y", InvoiceDate) AS YEAR,
COUNT (*) AS TOTAL_CLIENTS FROM invoices
WHERE (strftime("%Y", InvoiceDate) = "2009") OR (strftime("%Y", InvoiceDate) = "2011")
GROUP BY YEAR

'''

df = sql_query(query)
df
```

```python
# 13. Cuántas facturas ha habido entre 2009 y 2011

query = '''
SELECT strftime("%Y", InvoiceDate) AS YEAR, COUNT(*) AS TOTAL_CLIENTS  
FROM invoices  
WHERE strftime("%Y", InvoiceDate) BETWEEN "2009" AND "2011"  
GROUP BY YEAR;

'''

df = sql_query(query)
df
```

```python
# 1. Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la 
# factura y el país de la factura

query = '''
SELECT I.INVOICEID, I.INVOICEDATE, I.BILLINGCOUNTRY FROM INVOICES AS I
JOIN CUSTOMERS AS C ON C.CUSTOMERID = I.CUSTOMERID AND I.BillingCountry = "Brazil"
'''

df = sql_query(query)
df
```

```python
#  2. Obtén cada factura asociada a cada agente de ventas con su nombre completo

query = '''
SELECT * FROM INVOICES AS I
JOIN CUSTOMERS AS C ON C.CUSTOMERID = I.CUSTOMERID
JOIN EMPLOYEES AS E ON E.EMPLOYEEID = C.SUPPORTREPID
WHERE E.Title = "Sales Support Agent"
'''

df = sql_query(query)
df
```

```python
# 3. Obtén el nombre del cliente, el país, el nombre del agente y el total

query = '''
SELECT C.FIRSTNAME, C.COUNTRY, E.FIRSTNAME, I.TOTAL FROM CUSTOMERS AS C
JOIN EMPLOYEES AS E ON C.SUPPORTREPID = E.EmployeeId
JOIN INVOICES AS I ON C.CUSTOMERID = I.CUSTOMERID 
'''

df = sql_query(query)
df
```

```python
#  4. Obtén cada artículo de la factura con el nombre de la canción

query = '''
SELECT T.NAME, II.INVOICEID FROM INVOICE_ITEMS AS II
JOIN TRACKS AS T ON T.TRACKID = II.TrackId
'''

df = sql_query(query)
df
```

```python
#  5. Muestra todas las canciones con su nombre, formato, álbum y género
query = '''
SELECT DISTINCT T.NAME, A.TITLE, G.NAME, M.NAME FROM TRACKS AS T
JOIN ALBUMS AS A ON A.ALBUMID = T.ALBUMID
JOIN GENRES AS G ON G.GENREID = T.GENREID
JOIN MEDIA_TYPES AS M ON M.MediaTypeId = T.MEDIATYPEID
'''

df = sql_query(query)
df
```

```python
# 6. Cuántas canciones hay en cada playlist

query = '''
SELECT P.Name, COUNT(T.Name) AS CANTIDAD FROM TRACKS AS T
JOIN PLAYLIST_TRACK AS PT ON T.TrackId = PT.TrackId
JOIN PLAYLISTS AS P ON P.PlaylistId = PT.PlaylistId
GROUP BY P.Name
'''

df = sql_query(query)
df
```

```python
#  7. Cuánto ha vendido cada empleado
query = '''
SELECT E.FIRSTNAME, COUNT(I.INVOICEID) AS VENTAS FROM EMPLOYEES AS E
JOIN CUSTOMERS AS C ON C.SUPPORTREPID = E.EmployeeId
JOIN INVOICES AS I ON I.CUSTOMERID = C.CustomerId
GROUP BY E.FirstName

'''

df = sql_query(query)
df
```

```python
# 8. ¿Quién ha sido el agente de ventas que más ha vendido en 2009?

query = '''
SELECT E.FIRSTNAME, COUNT(I.INVOICEID) AS VENTAS FROM EMPLOYEES AS E
JOIN CUSTOMERS AS C ON C.SUPPORTREPID = E.EmployeeId
JOIN INVOICES AS I ON I.CUSTOMERID = C.CustomerId
WHERE CAST(STRFTIME("%Y", I.INVOICEDATE) AS INT) = 2009
GROUP BY E.FirstName
ORDER BY VENTAS DESC
LIMIT 1

'''

df = sql_query(query)
df
```

```python
#  9. ¿Cuáles son los 3 grupos que más han vendido?

query = '''
SELECT artists.Name AS Grupo,
       SUM(invoice_items.UnitPrice * invoice_items.Quantity) AS TotalVendido
FROM artists
JOIN albums ON artists.ArtistId = albums.ArtistId
JOIN tracks ON albums.AlbumId = tracks.AlbumId
JOIN invoice_items ON tracks.TrackId = invoice_items.TrackId
GROUP BY artists.ArtistId
ORDER BY TotalVendido DESC
LIMIT 3;

'''

df = sql_query(query)
df
```

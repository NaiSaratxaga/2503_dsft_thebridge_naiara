- Speed: La rapidez del Pokémon.
- Stage: Etapa evolutiva.
- Legendary: Indica si el Pokémon es legendario o no.

```python
# 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
# Dado un DataFrame, melt recibe varios parámetros:
# * qué columnas del DataFrame se quedan igual
# * qué nombres de columnas del dataframe pasan a ser una columna nueva
# * cómo queremos llamar esas nuevas columnas

# Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas

pokemon_type.head()

pokemon_extra=pokemon_type.iloc[:25,:]
pokemon_extra
```

```python
# melt() convierte columnas en filas.
# * qué columnas del DataFrame se quedan igual
# name y type1 se mantienen
#* qué nombres de columnas del dataframe pasan a ser una columna nueva
# - New Column= HP, Attack,Defense, Sp.Attack,Sp.Def y Speed 
# Value= los valores numéricos
# * cómo queremos llamar esas nuevas columnas
# var_name='Stat',  # Nueva columna que contendrá los nombres de las columnas originales 
# value_name='Value'  # Nueva columna que contendrá los valores correspondientes

pokemon_extra = pokemon_extra.melt(id_vars=['Name', 'Type 1'], value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk',   'Sp. Def', 'Speed'], var_name='New Column')
pokemon_extra
```

```python
# 20. Queremos dejar igual las variables 'Name' y 'Type 1'
# Queremos añadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
# Escribimos df_gfw.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')
```

```python
# 21. Pintamos el swarmplot con el melted y entendemos la información
plt.figure(figsize=(10, 6))

# swarmplot crea un gráfico de dispersión.
sns.swarmplot(data=pokemon_extra, x="New Column", y="value", hue="Type 1")

plt.legend(title="Tipo de Pokémon", loc='upper left', bbox_to_anchor=(1, 1))

# Título y etiquetas
plt.title("Distribución de Estadísticas de Pokémon por Tipo", fontsize=16)
plt.xlabel("Estadísticas", fontsize=12)
plt.ylabel("Valor de la Estadística", fontsize=12)
```

```python
# 22. Basándote en el gráfico anterior, define el tamaño de la figura en (10,6) (usa plt)
# En el swarmplot añade el argumento dodge a True y usa la anterior paleta de colores Pokemon

# Dodge en inglés es "esquivar", ¿qué diferencia ves?
# - "separa" los puntos de diferentes categorías dentro de una misma variable para que no se superpongan.
# - Si tienes varias categorías en el eje X (por ejemplo, diferentes tipos de Pokémon), los puntos de esas categorías se desplazan horizontalmente para evitar que se amontonen y se vean con claridad.

# Usando plt 
# fija el límite del eje Y entre 0 y 200
# fija la leyenda con bbox_to_anchor en upper left. Pista: upper left es el código 2 y se usa el parámetro loc.

plt.figure(figsize=(10, 6))
sns.swarmplot(data=pokemon_extra, x="New Column", y="value", hue="Type 1", dodge=True)
pokemon_palette = ['#78C850', '#F08030', '#6890F0', '#A8B820', '#A040A0', '#F8D030'] 

plt.legend(title="Tipo de Pokémon", loc='upper left', bbox_to_anchor=(1, 1))

# Establecer límites en el eje Y entre 0 y 200
plt.ylim(0, 200)

# Título y etiquetas
plt.title("Distribución de Estadísticas de Pokémon por Tipo", fontsize=16)
plt.xlabel("Estadísticas", fontsize=12)
plt.ylabel("Puntos", fontsize=12)
```

```python
# 23. Mapas de calor
# Calcula las correlaciones del primer DataFrame original df
# Pista: las correlaciones se calculan con el método corr()
# Recuerda: las correlaciones miden entre dos variables, si varío una, cuánto me varía la otra (cómo están de relacionadas)
```

```python

# Seleccionar solo las columnas numéricas, si no corr no funciona.
correlation_matrix = pokemon.select_dtypes(include=['number']).corr()

# annot= True, muestra los valores numéricos dentro de cada celda del heatmap.
# cmap='coolwarm', define el colormap (cmap) de la gráfica.
# fmt='.2f', Indica el formato de los números dentro de cada celda.
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True,vmin=-1,vmax=1, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Título del gráfico
plt.title("Matriz de Correlación del DataFrame", fontsize=16)


plt.show()
```

```python
# ¿la correlación entre qué valores se encuentra?
# la autocorrelación es la correlación de una variable consigo misma, ¿qué aprecias?
# Encuentra dos variables que veas que están fuertemente relacionadas según la correlación
# Si tuvieras una matriz de correlación muy grande y te cobrasen por cada dato que lees, ¿qué harías? (ML)
```

### ¿la correlación entre qué valores se encuentra?
- La correlación entre dos variables muestra el grado en el que ambas variables están relacionadas linealmente. La correlación en la matriz varía entre -1 y 1.
En este caso, los valores están entre -0.05 y 1.00, lo que indica correlaciones desde muy débiles (cercanas a 0) hasta perfectas (1.00).

### la autocorrelación es la correlación de una variable consigo misma, ¿qué aprecias?
- La autocorrelación (cuando una variable se correlaciona consigo misma) siempre es 1.00.
- La autocorrelación es de 1 (o muy cercana a 1) en la diagonal principal de la matriz de correlación, porque cualquier variable está perfectamente correlacionada consigo misma. 

### Encuentra dos variables que veas que están fuertemente relacionadas según la correlación
- Sp. Atk y Sp. Def tienen una correlación de 0.52, lo que indica una relación moderadamente fuerte.

Attack y Defense tienen una correlación de 0.49, lo que sugiere que Pokémon con alto ataque suelen tener también alta defensa.
### Si tuvieras una matriz de correlación muy grande y te cobrasen por cada dato que lees, ¿qué harías? (ML)
- Me centraría solo en las correlaciones fuertes y eliminaría aquellas que sean muy bajas o cercanas a cero, ya que no aportan información útil.

```python
# 24. Pinta el histograma y la densidad de probabilidad del nivel de defensa usando el DataFrame original

plt.figure(figsize=(10, 6))

# kde= True, activa la curva de densidad (KDE - Kernel Density Estimation).
# Esta curva suaviza el histograma para mostrar mejor la distribución de los datos.
# Es útil para ver si los datos siguen una distribución normal o si tienen sesgo.
sns.histplot(data=pokemon, x="Defense", bins=20, kde=True, color="skyblue")

# Etiquetas
plt.title("Histograma y Densidad de 'Defense'", fontsize=16)
plt.xlabel("Nivel de Defensa", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)


plt.show()

# Es ligeramente simétrica, es una distribución normal.
# La línea suaviza la forma del histograma y permite ver la distribución con más claridad.
```

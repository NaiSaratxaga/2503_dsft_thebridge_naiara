```python
# 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
# Dado un DataFrame, melt recibe varios parámetros:
# * qué columnas del DataFrame se quedan igual
# * qué nombres de columnas del dataframe pasan a ser una columna nueva
# * cómo queremos llamar esas nuevas columnas

# Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas

pokemon_type.head()

pokemon_extra=pokemon_type.iloc[:25,:]
pokemon_extra
```

```python
# melt() convierte columnas en filas.
# * qué columnas del DataFrame se quedan igual
# name y type1 se mantienen
#* qué nombres de columnas del dataframe pasan a ser una columna nueva
# - New Column= HP, Attack,Defense, Sp.Attack,Sp.Def y Speed 
# Value= los valores numéricos
# * cómo queremos llamar esas nuevas columnas
# var_name='Stat',  # Nueva columna que contendrá los nombres de las columnas originales 
# value_name='Value'  # Nueva columna que contendrá los valores correspondientes

pokemon_extra = pokemon_extra.melt(id_vars=['Name', 'Type 1'], value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk',   'Sp. Def', 'Speed'], var_name='New Column')
pokemon_extra
```

```python
# 20. Queremos dejar igual las variables 'Name' y 'Type 1'
# Queremos añadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
# Escribimos df_gfw.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')
```

```python
# 21. Pintamos el swarmplot con el melted y entendemos la información
plt.figure(figsize=(10, 6))
sns.swarmplot(data=pokemon_extra, x="New Column", y="value", hue="Type 1")

plt.legend(title="Tipo de Pokémon", loc='upper left', bbox_to_anchor=(1, 1))

# Título y etiquetas
plt.title("Distribución de Estadísticas de Pokémon por Tipo", fontsize=16)
plt.xlabel("Estadísticas", fontsize=12)
plt.ylabel("Valor de la Estadística", fontsize=12)
```

```python

# Seleccionar solo las columnas numéricas, si no corr no funciona.
correlation_matrix = pokemon.select_dtypes(include=['number']).corr()

# Crear un mapa de calor para visualizar la correlación
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Título del gráfico
plt.title("Matriz de Correlación del DataFrame", fontsize=16)

# Mostrar el gráfico
plt.show()
```

### ¿la correlación entre qué valores se encuentra?
- La correlación se calcula entre todas las variables numéricas de Pokémon, tales como HP, Attack, Defense, Sp. Atk, Sp. Def, Speed y Total.
- Específicamente, la correlación entre dos variables muestra el grado en el que ambas variables están relacionadas linealmente. Si el valor es cercano a 1, indica una fuerte correlación positiva (ambas variables aumentan juntas), si es cercano a -1, hay una correlación negativa (una variable aumenta mientras la otra disminuye), y si es cercano a 0, las variables no están relacionadas de manera lineal.

### la autocorrelación es la correlación de una variable consigo misma, ¿qué aprecias?
- La autocorrelación siempre será 1 (o muy cercana a 1) en la diagonal principal de la matriz de correlación, porque cualquier variable está perfectamente correlacionada consigo misma. Esto no es algo sorprendente, ya que una variable siempre tiene una relación perfecta (positiva) con ella misma. Por ejemplo:

Correlación de HP con HP es 1.

Correlación de Attack con Attack es 1.

Esto es lo que ves en la diagonal de la matriz de correlación.
### Encuentra dos variables que veas que están fuertemente relacionadas según la correlación
- Basándonos en los valores de correlación, por lo general, puedes ver que algunas variables están fuertemente relacionadas, como por ejemplo:

Attack y Sp. Atk: Estas dos variables suelen estar muy correlacionadas, ya que ambos se refieren a tipos de ataques, y los Pokémon que tienen un alto valor en uno suelen tener un alto valor en el otro.

Si la correlación entre estas dos variables es alta (cerca de 1), puedes concluir que Pokémon con un alto valor de Attack tienden a tener también un alto valor de Sp. Atk.

Defense y Sp. Def: A menudo, la defensa física (Defense) y la defensa especial (Sp. Def) están correlacionadas, ya que ambos valores se refieren a la capacidad de un Pokémon para resistir ataques, pero en diferentes formas (físicos vs. especiales).
### Si tuvieras una matriz de correlación muy grande y te cobrasen por cada dato que lees, ¿qué harías? (ML)
- Si tuviera una matriz de correlación muy grande y me cobraran por cada dato que leo, consideraría realizar las siguientes acciones:

Seleccionar solo las variables más importantes: En vez de calcular la correlación entre todas las variables, seleccionaría solo un subconjunto de las variables que son más relevantes para mi análisis. Esto podría ser un conjunto de variables con un mayor impacto en el modelo o las que tengan más valor en el contexto de la pregunta que estoy intentando responder.

Reducción de dimensiones: Podría usar técnicas como PCA (Análisis de Componentes Principales) para reducir la cantidad de variables, de manera que pueda analizar un conjunto más pequeño de variables que aún capturen la mayor parte de la información relevante de la matriz original. PCA permite reducir la dimensionalidad sin perder demasiada información importante.

Filtrar variables con baja correlación: Si estoy buscando relaciones significativas entre las variables, podría eliminar aquellas que tienen una correlación baja (cerca de 0). Esto reduciría el tamaño de la matriz de correlación y también evitaría trabajar con variables que no contribuyen mucho.

Uso de técnicas de selección de características (feature selection): Técnicas como SelectKBest o L1 regularization (usando modelos como regresión Lasso) pueden ayudar a identificar las características más relevantes de un conjunto de datos.

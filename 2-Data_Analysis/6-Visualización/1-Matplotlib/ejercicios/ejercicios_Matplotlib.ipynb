2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
# Crear valores del eje X entre 0 y 5 con precisión
x = np.linspace(0, 5, 100)

# Definir las ecuaciones de las rectas
y1 = 3 * x        # Recta con pendiente positiva (pasa por (0,0))
y2 = -3 * x - 5   # Recta con pendiente negativa (pasa por (0,-5))

# Con esto pintamos las 2 gráficas
# Crear la figura con dos subgráficos
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 fila, 2 columnas

# Gráfico de la izquierda (recta positiva)
axes[0].plot(x, y1, color='purple', label="y = 3x")
axes[0].set_title("RECTA POSITIVA")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].axhline(0, color='red', linewidth=0.5)  # Línea del eje X
axes[0].axvline(0, color='red', linewidth=0.5)  # Línea del eje Y
axes[0].legend()
axes[0].grid()

# Gráfico de la derecha (recta negativa)
axes[1].plot(x, y2, color='green', label="y = -3x - 5")
axes[1].set_title("RECTA NEGATIVA")
axes[1].set_xlabel("X")
axes[1].axhline(0, color='red', linewidth=0.5)
axes[1].axvline(0, color='red', linewidth=0.5)
axes[1].legend()
axes[1].grid()

# ajusta automáticamente los espacios entre los elementos de un gráfico 
# para evitar que se solapen o se vean mal distribuidos.
plt.tight_layout()
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python

# Crear el DataFrame con los datos de ventas
data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000] # columna nueva
}

df = pd.DataFrame(data)

# Configurar el estilo
plt.style.use('seaborn-v0_8-whitegrid')

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Ventas del Departamento A
ax.plot(df['Year'], df['Sold_items_A'], linestyle='--', color='green', label="Departamento A")

# Ventas del Departamento B
ax.plot(df['Year'], df['Sold_items_B'], linestyle=':', color='red', label="Departamento B")

# Añadir título y etiquetas
ax.set_title("Ventas de artículos por Año")
ax.set_xlabel("Año")
ax.set_ylabel("Ventas de artículos")

# Asegurar que solo aparecen los años del DataFrame en el eje X
ax.set_xticks(df['Year'])

# Agregar la leyenda abajo a la derecha
ax.legend(loc="lower right")

# Mostrar el gráfico
plt.show()
```
